# **Step 1: Import Required Libraries**
>Before starting the analysis, we import the required Python libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# visualization settings
sns.set_style("whitegrid")
```

**These libraries provide tools for:**

Data manipulation (Pandas)

Mathematical operations (NumPy)

Data visualization (Matplotlib and Seaborn)

# **Step 2: Loading the Dataset**

```python
df = pd.read_csv("movie_metadata.csv")

df.head()
```

This step loads the dataset into a **Pandas** DataFrame, which allows **structured manipulation** and **analysis of the data**.

# **Step 3: Dataset Exploration**

>Before cleaning the data, it is essential to understand its structure.

```python
df.info()
```

```python
df.describe()
```

**This helps identify:**

Data types

Missing values

Statistical summary of numerical columns

Dataset size and structure

# **Step 4: Data Cleaning and Preprocessing**

>Real-world datasets usually contain incomplete or inconsistent data.

```python
# Checking missing values
df.isnull().sum()
```

```python
# Remove rows without rating
df = df.dropna(subset=['imdb_score'])

# Remove duplicate records
df = df.drop_duplicates()
```

```python
# Convert year column to numeric
df['title_year'] = pd.to_numeric(df['title_year'], errors='coerce')
```

**This step ensures that the dataset becomes:**

Accurate

Consistent

Suitable for statistical analysis

**Step 5: Statistical Analysis of Movie Ratings**

>To understand rating behavior, we calculate descriptive statistics.

```python
mean_rating = df['imdb_score'].mean()
median_rating = df['imdb_score'].median()
mode_rating = df['imdb_score'].mode()[0]

print("Mean Rating:", mean_rating)
print("Median Rating:", median_rating)
print("Mode Rating:", mode_rating)
```

**Interpretation**

Mean Rating shows the average movie rating.

Median Rating indicates the middle value.

Mode Rating shows the most frequent rating.

>These statistics help understand overall audience sentiment.

# **Step 6: Rating Distribution Analysis**

**Histogram Visualization**

```python
plt.figure(figsize=(10,6))

sns.histplot(df['imdb_score'], bins=20, kde=True)

plt.title("Distribution of Movie Ratings")
plt.xlabel("IMDb Rating")
plt.ylabel("Frequency")

plt.show()
```

The histogram reveals how ratings are distributed across movies.

Typical observations include:

Most movies receive ratings between 6 and 8

Very few movies achieve extremely high ratings above 9

# **Step 7: Box Plot Analysis**

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df['imdb_score'])

plt.title("Box Plot of IMDb Ratings")

plt.show()
```

**The box plot helps identify:**

Rating spread

Median rating

Outliers (extremely high or low ratings)

# **Step 8: Top Rated Movies**

```python
top_movies = df.sort_values(by='imdb_score', ascending=False)

top_movies[['movie_title','imdb_score']].head(10)
```

This step identifies the highest rated movies in the dataset.

These films represent exceptional audience reception.

# **Step 9: Genre-Based Rating Analysis**

>Many movies belong to multiple genres. Therefore, we separate them.

```python
df['genres'] = df['genres'].str.split('|')

genre_df = df.explode('genres')

genre_ratings = genre_df.groupby('genres')['imdb_score'].mean()

genre_ratings = genre_ratings.sort_values(ascending=False)

genre_ratings.head(10)
```

**This analysis shows which genres consistently receive higher ratings.**

Often, genres such as:

Drama

Biography

Crime

tend to achieve higher audience appreciation.

# **Step 10: Genre Visualization**

```python
plt.figure(figsize=(12,6))

genre_ratings.head(10).plot(kind='bar')

plt.title("Top Genres Based on Average IMDb Ratings")
plt.xlabel("Genre")
plt.ylabel("Average Rating")

plt.show()
```

This chart visually compares genre performance.

# **Step 11: Analytical Dashboard**

>To summarize insights, we build a mini dashboard.

```python
fig, axes = plt.subplots(2,2, figsize=(14,10))

sns.histplot(df['imdb_score'], bins=20, ax=axes[0,0])
axes[0,0].set_title("Rating Distribution")

sns.boxplot(x=df['imdb_score'], ax=axes[0,1])
axes[0,1].set_title("Rating Spread")

genre_ratings.head(5).plot(kind='bar', ax=axes[1,0])
axes[1,0].set_title("Top Genres")

df.groupby('title_year')['imdb_score'].mean().plot(ax=axes[1,1])
axes[1,1].set_title("Ratings Over Years")

plt.tight_layout()

plt.show()
```

This dashboard provides a compact visual summary of the analysis.
